### Definition of a Metric for the Business to Monitor

#### Metric: **Average Revenue per Customer Interaction (ARCI)**

**Definition:**
Average Revenue per Customer Interaction (ARCI) measures the average revenue generated each time a sales representative interacts with a customer, regardless of the sales method used. It is calculated by dividing the total revenue by the total number of interactions.

$$
\text{ARCI} = \frac{\text{Total Revenue}}{\text{Total Number of Interactions}}
$$


This metric provides a clear indication of the effectiveness of the sales team's interactions with customers. By monitoring ARCI, the business can understand which sales methods are most effective in generating revenue and adjust their strategies accordingly.

### 1) How to use it to solve business problem? 

1. **Monitor Sales Method Effectiveness:**
   - Track ARCI for each sales method (Email, Call, Email + Call) to identify which method is generating the highest average revenue.
   - Use this information to allocate resources and prioritize sales methods that yield higher returns.

2. **Identify Trends Over Time:**
   - Analyze ARCI over different time periods (weekly, monthly) to identify trends and patterns.
   - Determine if certain times of the year or sales campaigns correlate with higher ARCI values.

3. **Set Performance Targets:**
   - Establish ARCI benchmarks based on historical data and set targets for the sales team.
   - Regularly review and adjust these targets to drive continuous improvement.

4. **Optimize Sales Strategies:**
   - Use ARCI data to refine sales approaches, such as increasing the use of methods with higher ARCI or experimenting with new combinations of sales tactics.
   - Implement training programs to improve the effectiveness of lower-performing methods.

# Appendix-1: Python script import

## Import Library


```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 

sales = pd.read_csv("product_sales.csv")
```

## Data validation & cleaning

```python
# I. Data validation 

# 1) Clean revenue column 
print(sales.revenue.mean())
print(sales.revenue.quantile(0.5))

# filling the NaN with median value, since the distribution of value is right skewed due to the outliers 
sales.revenue.fillna(sales.revenue.quantile(0.5), inplace = True)

sales.revenue.isna().sum()

# 2) clean the "sales method" column 
print(sales["sales_method"].value_counts()) # by doing so, we can find some repeated values such as: em + call, email, as a result, we can clean this column by regrouping these in 3 group: "Email", "Call", "Email + Call"

# step 1 : set conditions 
Email = sales["sales_method"].str.contains("^Email|^email") & ~sales["sales_method"].str.contains("Call")
Call = sales["sales_method"].str.contains("^Call")
Email_Call = sales["sales_method"].str.contains('\+')
conditions = [(Email), (Call), (Email_Call)]

# step 2: set category
categories = ["Email", "Call", "Email + Call"]

# step 3 : renaw the "sales_method" column values 
sales["sales_method"] = np.select(conditions, categories, default = "Others")

# 3) Clean the "year_as_customer" column 
sales["years_as_customer"].value_counts().reset_index().sort_values(by = "index", ascending = False)
# Since the company was founded in 1984, its age is 40 years old, as a result, the years_as_customer should be below or egale to 40 => as a result, we have to drop 2 rows : years_as_customer = 47 or 63
sales = sales.query("years_as_customer <= 40")

# 4) Check the quality of others columns 
sales["week"].value_counts()
sales["nb_sold"].value_counts()
sales["revenue"].value_counts()
sales["revenue"].isnull().sum()
len(sales["customer_id"].unique()) # check wether it's 1 uniquer client ID per row 
sales["nb_site_visits"].value_counts().reset_index().sort_values(by = "index")
sales["state"].value_counts().reset_index().sort_values(by = "index")
# conclusion: all of rest columns are cleans
```

### Single Variable Analysis

```python
def plot_subplots(sales):
    # Single variable analysis 

    # setting up for multiple subplots and adjusting the layout 
    fig, axes = plt.subplots(4, 2, figsize = (18, 18))
    plt.tight_layout(pad = 5.0)

    # The distribution of "week"
    sns.countplot(ax = axes[0,0], data = sales,  x = "week", palette = 'viridis')
    axes[0, 0].set_title('Distributiuon of sales by week')
    axes[0, 0].set_xlabel("Week")
    axes[0, 0].set_ylabel("Count")

    # The distribution of number of product sold
    sns.histplot(ax = axes[0, 1], data = sales['nb_sold'], kde = True, color = "orange")
    axes[0, 1].set_title('Distribution of Number Sold')
    axes[0, 1].set_xlabel('Number Sold')
    axes[0, 1].set_ylabel('Frequency')

    # The distribution of revenue 
    sns.histplot(ax=axes[1, 0], data=sales['revenue'], kde=True)
    axes[1, 0].set_title('Distribution of Revenue')
    axes[1, 0].set_xlabel('Revenue')
    axes[1, 0].set_ylabel('Frequency')

    # The distribution of years as customer
    sns.histplot(ax=axes[1, 1], data=sales['years_as_customer'], kde=True, color = "green")
    axes[1, 1].set_title('Distribution of Years as Customer')
    axes[1, 1].set_xlabel('Years as Customer')
    axes[1, 1].set_ylabel('Frequency')
    axes[1, 1].set_xlim(0, 40)

    # The distribution of number of site visits 
    sns.histplot(ax=axes[2, 0], data=sales['nb_site_visits'], kde=True, color = "blue")
    axes[2, 0].set_title('Distribution of Number of Site Visits')
    axes[2, 0].set_xlabel('Number of Site Visits')
    axes[2, 0].set_ylabel('Frequency')

    # The distribution of number of site visits 
    sns.boxplot(ax=axes[2, 1], data=sales, x= 'nb_site_visits', color = "skyblue")
    axes[2, 1].set_title('Distribution of Number of Site Visits')
    axes[2, 1].set_xlabel('Number of Site Visits')
    axes[2, 1].set_ylabel('Frequency')

    # Top 5 customer states
    customer_state = sales['state'].value_counts(normalize=True).reset_index().head(5)
    customer_state.columns = ['state', 'percentage']
    customer_state['frequency'] = round(customer_state['percentage'] * 100, 2)
    barplot = sns.barplot(ax=axes[3, 0], data = customer_state, y = "state", x = "frequency", hue = "state")
    for index, value in enumerate(customer_state['frequency']):
        barplot.text(value, index, f'{value}', color='black', ha="left", va="center")
    # axes[3, 0].set_xlim(0, max(customer_state['frequency']) + 200)
    axes[3, 0].set_title('Top 5 Customer States')
    axes[3, 0].set_xlabel('percentage of total population')
    axes[3, 0].set_ylabel('State')

    # The count of sales methods used 
    sales_methode = sales["sales_method"].value_counts().reset_index()
    sales_methode.columns = ["sales_method", "count"]
    barplot = sns.barplot(ax=axes[3, 1], data = sales_methode, x = "sales_method", y = "count", hue= "sales_method", palette= "muted")
    axes[3, 1].set_ylim(0, max(sales_methode['count']) + 2000)

    for p in barplot.patches:
        barplot.annotate(format(p.get_height(), '.1f'),
                         (p.get_x() + p.get_width() / 2., p.get_height()),
                         ha='center', va='center',
                         xytext=(0, 9),  # Adjust the vertical position of the text
                         textcoords='offset points')
    axes[3, 1].set_title('Count of Sales Methods Used ')
    axes[3, 1].set_ylabel("Count")
```
